```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("/content/drive/MyDrive/datasets/ml/boston_houses.csv")
df
```

```python
x=df.drop("MEDV",axis=1)
x
```

```python
y=df.MEDV.values
y
```

```python
cols=x.columns
cols
```

```python
rng=range(len(cols))
rng
```

```python
from sklearn.linear_model import ElasticNet
en=ElasticNet(l1_ratio=0.5)
model=en.fit(x,y)
```

```python
en.score(x,y)
```

```python
coef=en.coef_
coef
```

```python
import numpy as np
n=np.argmax(coef)
n
```

```python
cols[n]
```

```python
plt.plot(rng,coef)
plt.xticks(rng,cols)
```

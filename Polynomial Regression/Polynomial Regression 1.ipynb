```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('/content/drive/MyDrive/datasets/ml/Salary_Experience.csv')
df
```

```python
x=df.YearsExperience.values.reshape(-1,1)
x
```

```python
y=df.Salary.values
y
```

```python
plt.scatter(x,y)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x,y)
```

```python
sns.regplot(x=x,y=y,line_kws={'color':'red'})
```

```python
from sklearn.metrics import r2_score
r2_score(y,lr.predict(x))
```

```python
from sklearn.preprocessing import PolynomialFeatures
pf=PolynomialFeatures(degree=5)
x_poly=pf.fit_transform(x)
x_poly
```

```python
plr=LinearRegression()
plr.fit(x_poly,y)
```

```python
plt.scatter(x,y)
plt.plot(x,plr.predict(x_poly))
```

```python
r2_score(y,plr.predict(x_poly))
```

after degree 14 the score is not increasing rather it is decreasing

```python
for i in range(1,20):
  pf=PolynomialFeatures(degree=i)
  x_poly=pf.fit_transform(x)
  plr.fit(x_poly,y)
  print("degree ",i," score ",r2_score(y,plr.predict(x_poly)))
  plt.scatter(x,y)
  plt.plot(x,plr.predict(x_poly))
  plt.show()
  print("Polynomial Coefficients are ",plr.coef_)
  print("Intercept ",plr.intercept_)
  print("------------------------------------------------")
```

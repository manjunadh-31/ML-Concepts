```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("/content/drive/MyDrive/datasets/ml/Salary_Position.csv")
df
```

```python
x=df.iloc[:,1:2].values
x
```

```python
y=df.iloc[:,2].values
y
```

```python
plt.scatter(x=x,y=y)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x,y)
```

```python
y_pred=lr.predict(x)
y_pred
```

```python
from sklearn.metrics import r2_score
r2_score(y,y_pred)
```

```python
plt.scatter(x=x,y=y)
plt.plot(x,y_pred)
```

```python
from sklearn.preprocessing import PolynomialFeatures
pf=PolynomialFeatures(degree=9)
x_poly=pf.fit_transform(x)
x_poly
```

```python
plr=LinearRegression()
plr.fit(x_poly,y)
```

```python
plt.scatter(x=x,y=y)
plt.plot(x,plr.predict(x_poly))
```

```python
for i in range(1,17):
  pf=PolynomialFeatures(degree=i)
  x_poly=pf.fit_transform(x)
  plr.fit(x_poly,y)
  print("degree: ",i," score: ",r2_score(y,plr.predict(x_poly)))
  plt.scatter(x=x,y=y)
  plt.plot(x,plr.predict(x_poly))
  plt.show()
  print("---------------------------------")
```

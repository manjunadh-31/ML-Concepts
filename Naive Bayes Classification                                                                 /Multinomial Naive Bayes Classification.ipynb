# Multinimial Naive Bayes Classification

```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('/content/drive/MyDrive/datasets/ml/SMSSpamCollection',sep='\t',names=['status','msg'])
df
```

```python
df.shape
```

```python
df.status.value_counts()
```

```python
df.loc[df['status']=='spam','status']=0
df.loc[df['status']=='ham','status']=1
df
```

```python
x=df.msg.values
y=df.status
```

```python
x
```

```python
y.dtype
```

```python
y.astype('int')
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer()
xtraincv=cv.fit_transform(xtrain)
xtestcv=cv.transform(xtest)
```

```python
xtrain
```

```python
xtraincv.toarray()
```

```python
ytrain,ytest=ytrain.astype('int'),ytest.astype('int')
```

```python
from sklearn.naive_bayes import MultinomialNB
mnb=MultinomialNB()
mnb.fit(xtraincv,ytrain)
```

```python
mnb.score(xtraincv,ytrain)
```

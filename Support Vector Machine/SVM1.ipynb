```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_digits
data=load_digits()
data
```

```python
df=pd.DataFrame(data.data,columns=data.feature_names)
df
```

```python
df["target"]=data.target
df
```

```python
df.shape
```

```python
x=df.drop(["target"],axis=1)
x
```

```python
y=df.target
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=3)
```

```python
from sklearn.svm import SVC
sv=SVC(kernel="linear")
sv.fit(x_train,y_train)
display(sv.score(x_train,y_train))
sv.score(x_test,y_test)
```

```python
from sklearn.svm import SVC
sv=SVC(kernel="poly")
sv.fit(x_train,y_train)
display(sv.score(x_train,y_train))
sv.score(x_test,y_test)
```

```python
from sklearn.svm import SVC
sv=SVC(kernel="rbf")
sv.fit(x_train,y_train)
display(sv.score(x_train,y_train))
sv.score(x_test,y_test)
```

```python
from sklearn.decomposition import PCA

# Reduce to 2D using PCA
pca = PCA(n_components=2)
x_train_pca = pca.fit_transform(x_train)
x_test_pca = pca.transform(x_test)

# Train SVM on PCA-reduced data
sv_pca = SVC(kernel="linear")
sv_pca.fit(x_train_pca, y_train)

# Function to plot decision boundary
def plot_decision_boundary(model, X, y):
    h = 0.02  # Step size for mesh
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))

    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    plt.contourf(xx, yy, Z, alpha=0.3, cmap=plt.cm.coolwarm)
    scatter = plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.coolwarm, edgecolors="k")
    plt.xlabel("PCA Component 1")
    plt.ylabel("PCA Component 2")
    plt.title("SVM Decision Boundary (Linear Kernel)")
    plt.legend(*scatter.legend_elements(), title="Classes")
    plt.show()

plot_decision_boundary(sv_pca, x_train_pca, y_train)
```

```python
def plot_decision_boundary(model, X, y):
    h = 0.05  # Increase step size for speed
    X = X[:500]  # Use a subset of 500 points
    y = y[:500]

    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))

    # Reduce mesh grid size
    xx = xx[::5, ::5]
    yy = yy[::5, ::5]
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    plt.contourf(xx, yy, Z, alpha=0.3, cmap=plt.cm.coolwarm)
    scatter = plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.coolwarm, edgecolors="k")
    plt.xlabel("PCA Component 1")
    plt.ylabel("PCA Component 2")
    plt.title("Optimized SVM Decision Boundary (Linear Kernel)")
    plt.legend(*scatter.legend_elements(), title="Classes")
    plt.show()

plot_decision_boundary(sv_pca, x_train_pca, y_train)
```

```python
from sklearn.manifold import TSNE

# Reduce dimensions to 2D using t-SNE
tsne = TSNE(n_components=2, random_state=42)
x_train_tsne = tsne.fit_transform(x_train)

# Plot the data points
plt.figure(figsize=(8,6))
sns.scatterplot(x=x_train_tsne[:,0], y=x_train_tsne[:,1], hue=y_train, palette="coolwarm", edgecolor="k")
plt.xlabel("t-SNE Component 1")
plt.ylabel("t-SNE Component 2")
plt.title("t-SNE Visualization of Digits Dataset")
plt.legend(title="Digits")
plt.show()
```

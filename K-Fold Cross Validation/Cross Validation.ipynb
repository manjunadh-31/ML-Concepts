# **K-Fold Cross Validation**

K-Fold Cross-Validation is a technique used in machine learning to evaluate a modelâ€™s performance by splitting the dataset into K equal-sized subsets (folds). It helps to assess how well a model generalizes to unseen data.

How K-Fold Cross-Validation Works
1. **Split the Dataset:** The dataset is divided into K folds.

2. **Train & Test Multiple Times:**

 - The model is trained on K-1 folds.

 - The remaining 1 fold is used as the test set.

 - This process repeats K times, each time using a different fold as the test set.

3. **Compute Final Performance:** The final performance metric (e.g., accuracy, precision, recall) is the average of all K iterations.

```python
from sklearn import datasets
data=datasets.load_breast_cancer()
data
```

```python
x=data.data
y=data.target
```

```python
from sklearn.model_selection import KFold,cross_val_score
k=5
kf=KFold(n_splits=k)
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression(solver='liblinear')
cross_val_score(lr,x,y,cv=kf)
```

```python
cross_val_score(lr,x,y,cv=kf).mean()
```

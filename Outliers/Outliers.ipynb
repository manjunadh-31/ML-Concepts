3. IQR

```python
red.corr()
```

```python
sns.distplot(red["alcohol"])
```

```python
q1=red["alcohol"].quantile(0.25)
q1
```

```python
q3=red["alcohol"].quantile(0.75)
q3
```

```python
iqr=q3-q1
iqr
```

```python
ul=q3+1.5*iqr
ul
```

```python
ll=q1-1.5*iqr
ll
```

```python
upper=np.where(red["alcohol"]>ul)
upper
```

```python
lower=np.where(red["alcohol"]<ll)
lower
```

```python
empty =pd.DataFrame()
rows=[]
```

```python
for i in upper:
  print(red.iloc[i])
```

Eliminating Outliers

```python
red.drop(upper[0],inplace=True)
```

```python
red.drop(lower[0],inplace=True)
```

```python
upper=np.where(red["alcohol"]>ul)
upper
```

Replace outliers

```python
arr=red["alcohol"].values
arr
```

```python
true=(ll<arr) & (arr<ul)
true
```

```python
red["alcohol"][true]
```

```python
false=~(ll<arr) & (arr<ul)
false
```

```python
red["alcohol"][false]
```

```python
mid=red["alcohol"][true].median()
mid
```

```python
red["alcohol"].values[false]=mid
red["alcohol"].values[false]
```

```python
cols=red_df.columns
cols
```

```python
for col in cols:
    Q1 = red_df[col].quantile(0.25)
    Q3 = red_df[col].quantile(0.75)
    IQR = Q3 - Q1

    outlier_condition = (red_df[col] < (Q1 - 1.5 * IQR)) | (red_df[col] > (Q3 + 1.5 * IQR))

    median_value = red_df[col].median()

    red_df.loc[outlier_condition, col] = median_value
```

```python
def plot(i):
  sns.boxplot(red[i])
  plt.title(i)
  plt.show()
for col in cols:
  plot(col)
```

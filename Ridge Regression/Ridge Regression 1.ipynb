```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plr
```

```python
df=pd.read_csv("/content/drive/MyDrive/datasets/ml/boston_houses.csv")
df
```

```python
df.corr()
```

```python
df.info()
```

```python
x=df.RM.values.reshape(-1,1)
x
```

```python
x=df.iloc[:,5:6].values
x
```

```python
y=df.Price.values
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=5)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
lr.score(x_train,y_train)
```

```python
lr.score(x_test,y_test)
```

```python
import matplotlib.pyplot as plt

plt.scatter(x_train,y_train,marker="^")
plt.scatter(x_test,y_test,marker='+')
```

```python
plt.scatter(x_train,y_train,color="peachpuff",marker="^")
plt.plot(x_train,lr.predict(x_train),color="orange")
```

```python
from sklearn.linear_model import Ridge
rr=Ridge(alpha=0.01)
rr.fit(x_test,y_test)
```

```python
rr.score(x_test,y_test)
```

```python
plt.scatter(x_test,y_test,color="peachpuff")
plt.plot(x_test,rr.predict(x_test),color="orange")
```

```python
plt.scatter(x_train,y_train,color="peachpuff",marker="^")
plt.plot(x_train,lr.predict(x_train),color="orange")
plt.scatter(x_test,y_test,color="cyan",marker="+")
plt.plot(x_test,rr.predict(x_test),color="teal")
```

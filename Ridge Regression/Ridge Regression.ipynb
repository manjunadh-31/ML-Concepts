**Regularixation**

technique to minimize overfitting

Ridge regression, also known as  L2 regularization, is a technique used in linear regression to address the problem of multicollinearity among predictor variables. Multicollinearity occurs when independent variables in a regression model are highly correlated, which can lead to unreliable and unstable estimates of regression coefficients.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
x=pd.read_csv("/content/drive/MyDrive/datasets/ml/train.csv")
x
```

```python
y=pd.read_csv("/content/drive/MyDrive/datasets/ml/test.csv")
y
```

```python
w_x=x.drop(["Height"],axis=1).values
w_x
```

```python
h_x=x["Height"].values
h_x
```

```python
w_y=y.drop(["Height"],axis=1).values
w_y
```

```python
h_y=y["Height"].values
h_y
```

```python
plt.scatter(w_x,h_x,color="blue")
plt.scatter(w_y,h_y,color="red")
```

```python
  from sklearn.linear_model import LinearRegression
  lr=LinearRegression()
  lr.fit(w_x,h_x)
```

```python
sns.regplot(x=w_x,y=h_x,color="blue")
```

```python
lr.score(w_x,h_x)
```

```python
plt.scatter(w_x,h_x)
plt.plot(w_x,lr.predict(w_x),color="red")
```

```python
from sklearn.linear_model import Ridge
rr=Ridge(alpha=0.5)
rr.fit(w_y,h_y)
```

```python
rr.score(w_y,h_y)
```

```python
plt.scatter(w_y,h_y)
plt.plot(w_y,rr.predict(w_y),color="red")
```

```python
rr.predict([[70]])
```

```python
lr.predict([[70]])
```

# **Encoding**

It becomes easy for a model to understand the relationship between numeric typr of data.
Sometimes the categorical data will also contribute in final output which cannot be understood by the model. So the solution is to convert this data into numeric data using encoding techniques

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Dummies

convert the categorical data into binary number format using get_dummies()



```python
df=pd.read_csv("/content/drive/MyDrive/datasets/ml/homeprices.csv")
df
```

```python
dum=pd.get_dummies(df.town,dtype=int) #df["town"]
dum
```

```python
combi=pd.concat([df,dum],axis=1) #axis="columns"
combi
```

in below output it is possible to predict the 3rd value based on other values as all 3 values are having relationship hence they are not completely independent(dummy variable trap). To cut off this relationship we have to remove any one of the columns.


```python
x=combi.drop(["town","price","west windsor"],axis=1)
x
```

```python
y=df.price
y
```

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
reg.fit(x,y)
```

```python
from sklearn.metrics import r2_score
r2_score(y,reg.predict(x))
```

```python
reg.score(x,y)
```

```python
reg.predict([[2800,0,1]])
```

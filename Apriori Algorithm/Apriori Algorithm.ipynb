```python
lst=[]
for i in f:
  i=i.split(',')
  lst.append(i)
lst
```

```python
from apyori import apriori
res=apriori(lst,min_support=0.003,min_confidence=0.2,min_lift=4,min_length=2)
```

```python
f1=open('G:\My Drive\datasets\ml\Market_Basket__results.csv','w')
for i in res:
  f1.write(str(i)+"\n")
f1.close
```

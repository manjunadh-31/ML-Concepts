```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
pci=pd.read_excel('/content/drive/MyDrive/datasets/canada_per_capita_income.xlsx')
pci
```

```python
pci.describe()
```

```python
plt.figure(figsize=(15,10))
sns.barplot(x='year',y="income",data=pci)
plt.xticks(rotation=60)
```

```python
plt.figure(figsize=(15,10))
sns.lineplot(x='year',y="income",data=pci)
plt.xticks(rotation=60)
```

```python
plt.figure(figsize=(15,10))
sns.scatterplot(x='year',y="income",data=pci)
plt.xticks(rotation=60)
```

```python
pci.corr()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(pci[['year']],pci['income'],
                                               test_size=0.25,random_state=5)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
reg.fit(x_train,y_train)
```

```python
y_pred=reg.predict(x_test)
y_pred
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```

```python
reg.score(x_test,y_test)
```

# **Multiple Linear Regression**

y = β₀ + β₁x₁ + β₂x₂ + ... + βₙxₙ + ε.

y: The dependent variable.

x₁, x₂, ..., xₙ: The independent variables.

β₀: The intercept (the value of y when all x's are zero).

β₁, β₂, ..., βₙ: The regression coefficients, representing the change in y for a one-unit change in the corresponding x, holding other x's constant.

ε: The error term (representing the unexplained variance).

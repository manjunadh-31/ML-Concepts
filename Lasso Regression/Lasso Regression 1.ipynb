```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("/content/drive/MyDrive/datasets/ml/Advertising.csv")
df
```

```python
df.drop("Unnamed: 0",inplace=True,axis=1)
df
```

```python
x=df.drop("sales",axis=1)
x
```

```python
y=df["sales"].values
y
```

```python
from sklearn.linear_model import Lasso
lr=Lasso(alpha=0.01)
lr.fit(x,y)
```

```python
lr.score(x,y)
```

```python
cols=x.columns
cols
```

```python
rng=range(len(cols))
rng
```

```python
plt.plot(rng,lr.coef_)
plt.xticks(rng,cols,rotation="vertical")
```

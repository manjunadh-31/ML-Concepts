```python
sub_df=df[["satisfaction_level","average_montly_hours","promotion_last_5years","salary"]]
sub_df
```

```python
sal_dum=pd.get_dummies(df.salary,prefix="salary",dtype=int)
sal_dum
```

```python
new_df=pd.concat([sub_df,sal_dum],axis=1)
new_df
```

```python
new_df.drop(["salary","salary_medium"],inplace=True,axis=1)
new_df
```

```python
x=new_df
y=df.left
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)
```

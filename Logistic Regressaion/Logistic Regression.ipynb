# **Logistic Regression**

Logistic regression is a statistical method used primarily for binary classification problems where the outcome has two possible values (such as yes/no or success/failure).

 Unlike linear regression, which predicts continuous outcomes, logistic regression estimates the probability that a given input belongs to a particular category.

 The sigmoid function can be expressed mathematically as:

p=1/1 + e^(-z)

p is the predicted probability of the event occurring.

e is Euler's number (approximately 2.71828).

The linear combination z formed by multiplying each independent variable
xi by its coefficient βi and adding them together with the intercept β0 predicts the log-odds of the event occurring.


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("/content/drive/MyDrive/datasets/ml/insurance_data.csv")
df.sample(5)
```

```python
x=df.drop("bought_insurance",axis=1).values
x
```

```python
y=df.bought_insurance.values
y
```

```python
plt.scatter(x,y)
```

```python
sns.regplot(x=x,y=y,logistic=True)
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(x,y)
```

```python
lr.score(x,y)
```

```python
lr.predict([[56],[36]])
```
